```python
import os
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
graphs = sorted(os.listdir('learning_curves'))
graphs = [x for x in graphs if 'agrewards' in x]
print(graphs)
```

```python
for graph in graphs:
    with open(os.path.join('learning_curves', graph), 'rb') as f:
        data = pickle.load(f)
    data = data
    plt.plot(data)
    plt.title(graph[:-4])
    plt.show()
```
