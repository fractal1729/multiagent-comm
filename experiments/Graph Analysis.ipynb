```python
import os
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

## Per Agent Rewards

```python
graphs = sorted(os.listdir('learning_curves'))
graphs = [x for x in graphs if '_agrewards' in x]
print(graphs)
```

```python
for graph in graphs:
    with open(os.path.join('learning_curves', graph), 'rb') as f:
        data = pickle.load(f)
    data = np.asarray(data)
    for i in range(data.shape[1]):
        plt.plot(data[:,i])
    plt.title(graph[:-4].split('-')[0].replace('_', ' ') + " per agent reward")
    plt.show()
    break
```

## Episode Rewards

```python
graphs = sorted(os.listdir('learning_curves'))
graphs = [x for x in graphs if '_rewards' in x]
print(graphs)
```

```python
for graph in graphs:
    with open(os.path.join('learning_curves', graph), 'rb') as f:
        data = pickle.load(f)
    data = np.asarray(data)
    plt.plot(data)
    plt.title(graph[:-4].split('-')[0].replace('_', ' ') + " episode reward")
    plt.show()
```
